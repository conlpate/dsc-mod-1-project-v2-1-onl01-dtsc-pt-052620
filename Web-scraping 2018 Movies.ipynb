<b>General Imports for Web-Scraping</b>

```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
```

<b>Making Soup from Numbers</b>

```python
numbers_url = "https://www.the-numbers.com/market/2018/top-grossing-movies"
numbers_response = requests.get(numbers_url)
numbers_soup = BeautifulSoup(numbers_response.text, 'lxml')
```

Containers

```python
# Container with Table
numbers_warning = numbers_soup.find('table')

# Container with General Information
numbers_infocontainer = numbers_warning.find_all('tr')
```

Scraping Movies

```python
numbers_movies = []
a = 1
while a < len(numbers_infocontainer)-2:
    numbers_movies.append(numbers_infocontainer[a].find_all('a')[0].get_text())
    a += 1
print (len(numbers_movies))
for movie in numbers_movies:
    print (movie)
```

Scraping Release Date

```python
numbers_dates = []
b = 1
while b < len(numbers_infocontainer)-2:
    numbers_dates.append(numbers_infocontainer[b].find_all('a')[1].get_text())
    b += 1
print (len(numbers_dates))
for date in numbers_dates:
    print (date)
```

Scraping Genre

```python
print (numbers_infocontainer[1].find_all('a')[3].get_text())
print (numbers_infocontainer[2].find_all('a')[3].get_text())
print (numbers_infocontainer[3].find_all('a')[3].get_text())
print (numbers_infocontainer[736].find_all('a')[3].get_text())

numbers_genres = []
c = 1
while c < len(numbers_infocontainer)-2:
    numbers_genres.append(numbers_infocontainer[c].find_all('a')[1].get_text())
    c += 1
print (len(numbers_genres))
for genre in numbers_genres:
    print (genre)
```

Scraping Gross

```python
numbers_gross = []
d = 1
while d < len(numbers_infocontainer)-2:
    numbers_gross.append(numbers_infocontainer[d].find_all('td', class_="data")[1].get_text())
    d += 1
print (len(numbers_gross))
for gross in numbers_gross:
    print (gross)
```

Pandas

```python
numbers_df = pd.DataFrame([numbers_dates, numbers_genres, numbers_movies, numbers_gross]).transpose()
numbers_df.columns = ['Release_Date', 'Genre', 'Movie_Title', 'Domestic_Gross']
numbers_df
```

<b>Making Soup from IMDB (1)</b>

```python
imdb_url = "https://www.imdb.com/search/title/?title_type=feature&year=2018-01-01,2018-12-31&sort=boxoffice_gross_us,desc"
imdb_response = requests.get(imdb_url)
imdb_soup = BeautifulSoup(imdb_response.text, 'lxml')
```

Scraping Movies

```python
# Container with Target Space
imdb_warning = imdb_soup.find('div', class_="lister list detail sub-list")

# Container with General Information
imdb_infocontainer = imdb_warning.find('div', class_="lister-list")

# Container Strictly for Movie Names
imdb_moviescontainer = imdb_infocontainer.find_all('div', class_="lister-item-content")

print (len(imdb_moviescontainer))
imdb_movies = []
i = 0
while i < len(imdb_moviescontainer):
    imdb_movies.append(imdb_moviescontainer[i].find('a'))
    i += 1
for movie in imdb_movies:
    print(movie.get_text())
```

Scraping Genres

```python
imdb_genres = imdb_infocontainer.find_all('span', class_="genre")

print (len(imdb_genres))
for genre in imdb_genres:
    print(genre.text)
```

Scraping Gross

```python
imdb_gross = imdb_infocontainer.find_all('span', attrs={"name": "nv"})
real_imdb_gross = []
j = 0
while j < len(imdb_gross):
    if j%2 == 1:
        real_imdb_gross.append(imdb_gross[j])
    j += 1

print (len(real_imdb_gross))
for gross in real_imdb_gross:
    print (gross.get_text())
```

Pandas

```python
imdb_dates = []
imdb_df = pd.DataFrame([imdb_dates, imdb_genres, imdb_movies, real_imdb_gross]).transpose()
imdb_df.columns = ['Release_Date', 'Genre', 'Movie_Title', 'Domestic_Gross']
imdb_df
```
