<b>46. Seasonal Plot</b>
<br>Release Date VS ROI

```python
#46. Seasonal Plot
from dateutil.parser import parse 

# Import Data
df = pd.read_csv('MoviesDataFrame.csv')

# Prepare data
df['year'] = [parse(d).year for d in df.release_date]
df['month'] = [parse(d).strftime('%b') for d in df.release_date]
years = df['year'].unique()

# Draw Plot
mycolors = ['tab:red', 'tab:blue', 'tab:green', 'tab:orange', 'tab:brown', 'tab:grey', 'tab:pink', 'tab:olive', 'deeppink', 'steelblue', 'firebrick', 'mediumseagreen']      
plt.figure(figsize=(16,10), dpi= 80)

for i, y in enumerate(years):
    plt.plot('month', 'roi', data=df.loc[df.year==y, :], color=mycolors[i], label=y)
    plt.text(df.loc[df.year==y, :].shape[0]-.9, df.loc[df.year==y, 'roi'][-1:].values[0], y, fontsize=12, color=mycolors[i])

# Decoration
plt.ylim(100,50000)
plt.xlim(-0.3, 11)
plt.ylabel('$ROI$')
plt.yticks(fontsize=12, alpha=.7)
plt.title("Monthly Seasonal Plot: Movie Release ROI (2010 - 2019)", fontsize=22)
plt.grid(axis='y', alpha=.3)

# Remove borders
plt.gca().spines["top"].set_alpha(0.0)    
plt.gca().spines["bottom"].set_alpha(0.5)
plt.gca().spines["right"].set_alpha(0.0)    
plt.gca().spines["left"].set_alpha(0.5)   
# plt.legend(loc='upper right', ncol=2, fontsize=12)
plt.show()
```
