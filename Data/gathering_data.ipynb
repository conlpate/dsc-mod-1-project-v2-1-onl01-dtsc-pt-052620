Importing Necessary Packages

```python
import pandas as pd
```

Loading DataFrames from Files

```python
df1 = pd.read_csv('bom.movie_gross.csv.gz')
df1
```

```python
df2 = pd.read_csv('imdb.title.basics.csv.gz')
df2
```

```python
df3 = pd.read_csv('tn.movie_budgets.csv.gz')
df3
```

Editting Individual DataFrames for Target Data

```python
df1.drop(['studio'], axis=1, inplace=True)
df1.head()
```

```python
df2.drop(['tconst', 'original_title', 'start_year', 'runtime_minutes'], axis=1, inplace=True)
df2 = df2.rename(columns={'primary_title': 'title'})
df2.head()
```

```python
df3.drop(['id'], axis=1, inplace=True)
df3 = df3.rename(columns={'movie': 'title'})
df3.head()
```

Merging DataFrames

```python
merged_df1 = df1.merge(df2, on='title', how='left')
merged_df1
```

```python
merged_df1.info()
```

```python
merged_df2 = merged_df1.merge(df3, on='title', how='left')
merged_df2.head()
```

```python
merged_df2.info()
```

Editing Columns

```python
merged_df2.columns
```

```python
merged_df2['domestic_gross_x'].isnull
merged_df2['foreign_gross'].isnull
merged_df2['year'].isnull
merged_df2['genres'].isnull
merged_df2['release_date'].isnull
merged_df2['production_budget'].isnull
merged_df2['domestic_gross_y'].isnull
merged_df2['worldwide_gross'].isnull
```

```python
merged_df2['domestic_gross'] = merged_df2['domestic_gross_x'].map('{:,.2f}'.format)
merged_df2.head()
```

```python
merged_df2.drop(['domestic_gross_x', 'domestic_gross_y'], axis=1, inplace=True)
merged_df2.columns
```

```python
columnsorder = ['year', 'release_date', 'title', 'genres', 'production_budget', 'domestic_gross', 'foreign_gross',
                'worldwide_gross']
final_df = merged_df2.reindex(columns=columnsorder)
final_df.head()
```

```python
presentation_df = final_df.rename(columns={'release_date': 'Release Date', 'production_budget': 'Production Budget', 
                                           'domestic_gross': 'Domestic Gross', 'foreign_gross': 'Foreign Gross', 
                                           'worldwide_gross': 'Worldwide Gross'})
presentation_df.columns = presentation_df.columns.str.title()
presentation_df.head()
```
